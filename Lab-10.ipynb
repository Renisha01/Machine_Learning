# Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
```

# Step 2: Generate dummy data using make_blobs

```python
data, labels = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

# Step 3: Plot Generated data 

```python
plt.scatter(data[:, 0], data[:, 1], c='blue', edgecolors='k')
plt.title("Generated Data (Blobs)")
plt.show()
```

# Step 4: Apply Hierarchical Clustering

```python
hierarchical = AgglomerativeClustering(n_clusters=3)
```

```python
pred_labels = hierarchical.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=pred_labels, cmap='rainbow', edgecolors='k')
plt.title("Hierarchical Clustering")
plt.show()
```

# Mall Customers Segmentation using DBSCAN

### Perform Following steps on Dataset for  customer segmentation
-> Load & Explore the Dataset <br>
-> Data Preprocessing <br>
-> Apply DBSCAN <br>
-> Visualizing Customer Segments <br>
-> Outliers (high-spending VIP customers or unusual shoppers <br>

```python
df = pd.read_csv("Mall_Customers.csv")
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
dbscan_customers = DBSCAN(eps=5, min_samples=5).fit(X)
df['Cluster'] = dbscan_customers.labels_
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='rainbow')
plt.title("DBSCAN on Mall Customers")
plt.show()
```
