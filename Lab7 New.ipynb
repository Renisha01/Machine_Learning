# Do all pre processing on diabetes csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

# Import diabetes.csv data ser using sklearn

```python
df = pd.read_csv("diabetes.csv")
```

# Check the distribution of the target

```python
print(df['Outcome'].value_counts())
```

# Replace zeros with NaN for specific columns

```python
colums_to_replace = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[colums_to_replace] = df[colums_to_replace].replace(0, np.nan)
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Fill missing values with median

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
print(df.isnull().sum())
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({"Diabetic":1,"Non Diabetic":0})
```

```python
df
```

# Divide the data into input and output

```python
x = df.iloc[::,:-1:]
x
```

```python
y = df["Outcome"]
```

# import KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier 
```

# Convert all features into Z score

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_transform = scaler.fit_transform(x)
x_transform
```

# Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x_transform, y, test_size = 0.3, random_state=0)
```

# Create KNN Model

```python
model = KNeighborsClassifier(weights='uniform') #n ni value jetlo training data aipo che ena square-root thi check karvanu start karvanu
```

# Fitting KNN   on dataset

```python
model.fit(x_train,y_train)
```

# Display classification_report

```python
from sklearn.metrics import classification_report
```

```python
#print(classification_report(y_true, y_pred, target_names=target_names))
print(classification_report(y_test,model.predict(x_test)))
```

# Use Cross Validation to find best hyperparameter for KNN

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5)
```

```python
for i, (train_index, test_index) in enumerate(kf.split(x_transform,y)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")
```

```python
scores = cross_val_score(model, x_transform, y, cv=kf)
scores
```

```python
from sklearn.model_selection import GridSearchCV
```
